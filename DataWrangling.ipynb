```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```python
dem_results = {}
gop_results = {}
for state in states:
    url = 'https://townhall.com/election/2016/president/'+state+'/county'
    page = requests.get(url)
    soup = BeautifulSoup(page.text,'lxml')
    dem_results[state] = int(soup.findAll('td',{'class':'DEM'})[1].contents[0].replace(',',''))
    gop_results[state] = int(soup.findAll('td',{'class':'GOP'})[1].contents[0].replace(',',''))
```
